## 2. company_employee_number

```python
employ_df = sample[['estimated_number_employees_iv', 'employees_on_linkedin_oxy', 'estimated_current_number_of_emplyees_pdl',
                    'total_estimated_number_of_employees_pdl', 'company_employee_number']]
employ_df
```

```python
cols = ['estimated_number_employees_iv', 'employees_on_linkedin_oxy', 'estimated_current_number_of_emplyees_pdl',
                    'total_estimated_number_of_employees_pdl']
employ_df[cols].replace({0:np.nan}).info()
```

```python
print('We have no employment data for {:.2%} of the companies'.
      format(len(employ_df[employ_df['company_employee_number']=='[0.0]'])/len(employ_df)))
```

## 3. emp_range

```python
emprange_df = sample[['number_of_employees_oxy', 'estimated_range_of_employees_pdl', 'company_employee_range']]
emprange_df.info()
```

```python
emprange_df.dropna(how='all', subset=['number_of_employees_oxy', 'estimated_range_of_employees_pdl']).info()
```

```python
emprange_df[(emprange_df['number_of_employees_oxy'].isnull()) & (emprange_df['estimated_range_of_employees_pdl']).isnull()]
```

```python
emprange_df['diff_vals'] =  emprange_df['company_employee_range'].apply(lambda x: x.count(','))
print('Companies whose range from the 2 sources do not match: {:.2%}'.format(emprange_df['diff_vals'].sum()/len(emprange_df)))
```

## 5. company_industry

```python
ind_df = sample[['company_industry_oxy' ,'company_industry_pdl', 'company_industry_iv', 'company_industry']]
ind_df
```

```python
ind_df.info()
```

```python
ind_df['ind_count'] =  ind_df['company_industry'].apply(lambda x: x.count(',')-x.count("''")+1)
print('Companies with different industries : {:.2%}'.format(len(ind_df[ind_df['ind_count']>1])/len(ind_df)))
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    display(ind_df[(ind_df['ind_count']>1) & (ind_df['company_industry_iv'].notnull())])
```
